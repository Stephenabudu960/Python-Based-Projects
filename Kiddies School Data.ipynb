```python
import pandas as pd
import numpy as np
from faker import Faker
import random

# Initialize faker and seed for reproducibility
fake = Faker()
Faker.seed(42)
np.random.seed(42)

# Settings
num_rows = 500

# Possible categories
genders = ['Male', 'Female']
grade_levels = ['Nursery 1', 'Nursery 2', 'KG', 'Primary 1', 'Primary 2', 'Primary 3', 'Primary 4']
remarks = ['Excellent', 'Good', 'Satisfactory', 'Needs Improvement', 'Outstanding']

# Generate data
data = {
    'student_id': [f'STU{str(i).zfill(4)}' for i in range(1, num_rows + 1)],
    'name': [fake.first_name() + " " + fake.last_name() for _ in range(num_rows)],
    'gender': [random.choice(genders) for _ in range(num_rows)],
    'age': [random.randint(3, 10) for _ in range(num_rows)],
    'grade_level': [random.choice(grade_levels) for _ in range(num_rows)],
    'math_score': [random.randint(40, 100) for _ in range(num_rows)],
    'english_score': [random.randint(40, 100) for _ in range(num_rows)],
    'science_score': [random.randint(40, 100) for _ in range(num_rows)],
    'attendance_rate': [round(random.uniform(80.0, 100.0), 2) for _ in range(num_rows)],
    'teacher_remark': [random.choice(remarks) for _ in range(num_rows)],
}

# Create DataFrame
df = pd.DataFrame(data)

# Preview
df.head()

df.to_csv("kiddies_school_dataset.csv", index=False)
```
